### Loading asx tables to dataframe

```python
from IPython.display import display, HTML
mydb = connect_mysql(host, id, pwd, db)

#asx_all table for GICS sector details
asx_all_query = pd.read_sql_query("select * from asx.asx_all", con=mydb)
asx_all_df = pd.DataFrame(asx_all_query, columns=['Code','Company name','GICS industry group'])

#asx_history for 2020 (monthly data)
asx_history_query = pd.read_sql_query("select * from asx.asx_history where Date > '2020-01-01'", con=mydb)
asx_history_df = pd.DataFrame(asx_history_query, columns=['Code','Date','Open','High','Low','Close','Adj Close','Volume'])

#joining
asx_sector_price = pd.merge(asx_all_df, asx_history_df, on="Code")
asx_sector_price.head(10)
```

### Exploration and analysis

```python
#what are the unique sector values? 
list(asx_sector_price['GICS industry group'].unique())
```

## How many stocks are traded per sector?
#### Materials, Real-estate, Software&services, Diversified financials and Consumer services trade the most stocks

```python
#how many stocks are traded per sector?
count_of_stocks= asx_sector_price.groupby('GICS industry group')['Code'].nunique().sort_values().reset_index(name="count")
count_of_stocks.plot(x="GICS industry group", y="count",kind='barh')
```

## Which sectors have the highest volume traded? 

### Materials and Realestate have the highest volume. Energy, transportation and Banks, though trade less stocks comparatively have higher volumes.

```python
#sectors with highest volume traded 
total_volume_sector = asx_sector_price.groupby('GICS industry group')['Volume'].agg('sum').sort_values().reset_index(name="total_volume")
total_volume_sector.plot(x="GICS industry group", y="total_volume",kind='barh')

```

```python
#trend analysis - how did the sectors trade (in volume) in 2020
monthly_total_volume_sector = asx_sector_price.groupby(['GICS industry group', 'Date'])['Volume'].agg('sum').reset_index(name="total_volume")

fig = px.line(monthly_total_volume_sector, x="Date", y="total_volume", color='GICS industry group')
fig.show()
```

```python
#now, let's drill down to some inetresting sectors to see which underlying stocks contribute to the movements 
#materials
monthly_total_volume_company = asx_sector_price.groupby(['GICS industry group', 'Date', 'Company name'])['Volume'].agg('sum').reset_index(name="total_volume")

monthly_total_volume_company_MATERIAL = monthly_total_volume_company[(monthly_total_volume_company['GICS industry group'] == 'Materials')]

monthly_total_volume_company_MATERIAL_topStocks = monthly_total_volume_company_MATERIAL.groupby('Company name')['total_volume'].agg('sum').sort_values().tail(7)
monthly_total_volume_company_MATERIAL_topStocks.plot(x="Company name", y="total_volume",kind='barh')


fig1 = px.line(monthly_total_volume_company_MATERIAL, x="Date", y="total_volume", color='Company name')
fig1.show()

```
