```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# Import dataset and rename as `gdp_usd`
gdp_usd = pd.read_csv('australia_GDP_1960_2020_usd.csv')
gdp_usd.columns = ['date', 'gdp_billion','per_capita']
gdp_usd.head()
```

```python
# Convert data type of 'date' column to date. So that the dataframe could be ordered by date in the following steps. 
gdp_usd['date'] = pd.to_datetime(gdp_usd['date'], format = '%d/%m/%Y')
```

```python
# Briefly check dataset
gdp_usd.info()
gdp_usd.describe()
```

```python
sns.distplot(gdp_usd['gdp_billion'], bins=8) 

sns.distplot(gdp_usd['gdp_billion'], kde = True , bins = 30)
```

```python
sns.distplot(gdp_usd['gdp_billion']) 

sns.distplot(gdp_usd['gdp_billion'])
```

```python
sns.distplot(gdp_usd['per_capita'], bins=8) 

sns.distplot(gdp_usd['per_capita'], kde = True , bins = 30)
```

```python
sns.distplot(gdp_usd['per_capita']) 

sns.distplot(gdp_usd['per_capita'])
```

```python
gdp_billion = gdp_usd['gdp_billion'].tolist()
per_capita = gdp_usd['per_capita'].tolist()
date = gdp_usd['date'].tolist()

plt.plot(date, gdp_billion)
plt.title('GDP Vs Date')
plt.xlabel('Quarterly GDP')
plt.ylabel('GDP in USD (Billion)')
plt.show()
```

```python
gdp_billion = gdp_usd['gdp_billion'].tolist()
per_capita = gdp_usd['per_capita'].tolist()
date = gdp_usd['date'].tolist()

plt.plot(date, per_capita)
plt.title('GDP per_capita')
plt.xlabel('Year')
plt.ylabel('GDP per_capita in USD')
plt.show()
```

```python
gdp_billion = gdp_usd['gdp_billion'].tolist()
per_capita = gdp_usd['per_capita'].tolist()
date = gdp_usd['date'].tolist()

fig = plt.figure(figsize = (16,10), dpi = 300)

axes1 = fig.add_subplot(111)
axes1.set_xlabel('date')
axes1.set_ylabel('per_capita in USD')
lns1 = axes1.plot(date, per_capita, marker = ".", color = "red", label = "GDP per_capita in USD")
axes1.tick_params(axis='y', labelcolor="red")
axes1.legend(loc = "upper left")

axes2 = axes1.twinx()

color = 'tab:purple'
axes2.set_ylabel('Quarterly GDP in USD (Billion)')
lns2 = axes2.plot(date, gdp_billion, marker = ".", color = color, label = 'GDP in USD (Billion)')

plt.title('Australian GDP Historical Data')

lns = lns1+lns2
labs = [l.get_label() for l in lns]
axes1.legend(lns, labs, loc="upper left")

plt.show()
```

```python
# Import dataset and rename as `gdp_growth`
gdp_growth = pd.read_csv('australia_GDP_growth_1961_2020.csv')
gdp_growth.columns = ['date', 'gdp_growth','annual_change']
gdp_growth.head()
```

```python
# Convert data type of 'date' column to date. So that the dataframe could be ordered by date in the following steps. 
gdp_growth['date'] = pd.to_datetime(gdp_growth['date'], format = '%d/%m/%Y')
```

```python
# Briefly check dataset
gdp_growth.info()
gdp_growth.describe()
```

```python
sns.distplot(gdp_growth['gdp_growth'], bins=8)

sns.distplot(gdp_growth['gdp_growth'], kde = True , bins = 30)
```

```python
sns.distplot(gdp_growth['gdp_growth'])

sns.distplot(gdp_growth['gdp_growth'])
```

```python
gdp_annual_growth = gdp_growth['gdp_growth'].tolist()
differences = gdp_growth['annual_change'].tolist()
date = gdp_growth['date'].tolist()

plt.plot(date, gdp_annual_growth)
plt.title('GDP Growth Rate')
plt.xlabel('Quarter')
plt.ylabel('GDP Growth Rate')
plt.show()
```

```python
gdp_annual_growth = gdp_growth['gdp_growth'].tolist()
differences = gdp_growth['annual_change'].tolist()
date = gdp_growth['date'].tolist()

plt.plot(date, differences)
plt.title('GDP Growth Rate in Percentage')
plt.xlabel('Quarter')
plt.ylabel('Annual Changes')
plt.show()
```

```python
gdp_annual_growth = gdp_growth['gdp_growth'].tolist()
differences = gdp_growth['annual_change'].tolist()
date = gdp_growth['date'].tolist()

plt.figure(figsize = (16,10), dpi = 300)

plt.plot(date, gdp_annual_growth, label = "Annual GDP Growth")
plt.plot(date, differences, label = "Annual Differences")
plt.legend(loc="upper right")
plt.title('Australian GDP Annual Growth Historical Data')
plt.xlabel('Date')
plt.ylabel('Annual Growth Rate')
plt.show()
```

```python
# Import dataset and rename as `house_commitment`
house_commitment = pd.read_csv('new_loan_commitment_value_properties purchased.csv')
house_commitment = house_commitment.iloc[:,0:5]
house_commitment.head()
```

```python
# Convert data type of 'date' column to date. So that the dataframe could be ordered by date in the following steps. 
house_commitment['Date'] = pd.to_datetime(house_commitment['Date'], format = '%b-%y')
```

```python
# Briefly check dataset
house_commitment.info()
house_commitment.describe()
```

```python
total = house_commitment['Total New Loan Commitment Value of Properties Purchased'].tolist()
oo = house_commitment['New Loan Commitment Value of Owner Occupied Properties Purchased'].tolist()
inv = house_commitment['New Loan Commitment Value of Investment Properties Purchased'].tolist()
fhb = house_commitment['New Loan Commitment Value of Owner Occupied Properties Purchased (First Home Buyer)'].tolist()
month = house_commitment['Date'].tolist()

fig = plt.figure(figsize = (16,10), dpi = 300)

plt.plot(month, total, label = "Total")
plt.plot(month, oo, label = "Owner Occupied")
plt.plot(month, inv, label = "Investment")
plt.plot(month, fhb, label = "First Home Buyer")
plt.legend(loc="upper left")
plt.title('Australian Housing Commitments')
plt.xlabel('Date')
plt.ylabel('Amount of New Loan Commitment in AUD(millions)')
plt.show()
```
